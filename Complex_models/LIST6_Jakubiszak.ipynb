```
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```
def bass_model(p,q,population_size,stop_time):

    population = np.zeros(population_size)
    time = range(stop_time)

    innovators = np.zeros(stop_time)
    imitators = np.zeros(stop_time)

    adopters=0
    
    for t in time:
        for i in range(len(population)):
            x = np.random.uniform()

            if x < p*(1-population[i]):
                population[i] = 1
                innovators[t] += 1
            
            elif x<(p+q*adopters/population_size)*(1-population[i]):
                population[i] = 1
                imitators[t] += 1

        adopters += innovators[t]+imitators[t]

    return time,innovators,imitators
```

Below are the simulations of the bass diffusion model for the different sets of parameters.

```
population_size=1000
stop_time=30
p=0.03
q=0.3
```

```
time1,inn1,imi1=bass_model(p,q,population_size,stop_time)

for i in range(99):
    time,inn,imi=bass_model(p,q,population_size,stop_time)
    time1 = [time1[t]+time[t] for t in range(len(time1))]
    inn1 = [inn1[i]+inn[i] for i in range(len(inn1))]
    imi1 = [imi1[i]+imi[i] for i in range(len(imi1))]

time=[t/100 for t in time1]
inn=[i/100 for i in inn1]
imi=[i/100 for i in imi1]
```

```
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,9))


ax1.plot(time,np.cumsum(np.array(imi)+np.array(inn)))
ax1.set_title('Adopters')


ax2.plot(time,imi,label="Imitators")
ax2.plot(time,inn,label="Innovators")
ax2.plot(time,np.array(imi)+np.array(inn),label="New adopters")
ax2.set_title('New adopters')
ax2.legend()
f.show()
```

We can observe the simmilar behaviour as in the wikipedia article for p=0.03 and q=0.3
